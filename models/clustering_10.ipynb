```
import pandas as pd
from sklearn.cluster import KMeans
import glob
import pandas as pd
import numpy as np
```

```
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Bike_sharing/cleaned_csv/normed_data.csv')
```

```
data.set_index(data['Start station number'], drop=True, inplace=True)
```

```
data
```

```
data.drop('Start station number', axis=1, inplace=True)
```

```
data
```

```
X = data.to_numpy()
```

```
X
```

```
kmeans = KMeans(n_clusters=10, random_state=0).fit(X)
```

```
kmeans.labels_
```

```
data['cluster'] = kmeans.labels_
```

```
data
```

```
def increase_priority(num):
  station_dict = {
      0:9,
      1:8,
      2:7,
      3:6,
      4:5,
      5:4,
      6:3,
      7:2,
      8:1,
      9:0
  }
  return station_dict.get(num)
```

```
data['cluster'] = data['cluster'].apply(increase_priority)
```

```
group_length = [0] * 10
for val in enumerate(data['cluster']):
  group_length[val[1]] = group_length[val[1]] + 1
print(group_length)
group_length_dict = {}
for i, val in enumerate(group_length):
  group_length_dict.update({i:val})
print(group_length_dict)
```

```
data
```

```
cluster_dict = []
for index, row in data.iterrows():
  simple_dict = {"station_id" : index, "cluster_id" : row['cluster']}
  cluster_dict.append(simple_dict)
print(cluster_dict)
```

```
import csv
csv_file = "/content/drive/My Drive/Colab Notebooks/Bike_sharing/cleaned_csv/stations_cluster_mapping.csv"
csv_columns = ['station_id', 'cluster_id']
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for station in cluster_dict:
          writer.writerow(station)
except IOError:
    print("I/O error")
```

```
trips = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Bike_sharing/cleaned_csv/trips.csv')
```

```
trips = trips.drop(["Start date", "End station number", "day",  "workingday", "month", "time_of_day"], axis=1)
```

```
def group_stations(num):
   return cluster_dict.get(num)
```

```
trips['grouped_station'] = trips['Start station number'].apply(group_stations)
```

```
trips
```

```
trips = trips.drop(["Start station number"], axis=1)
```

```
trips = trips.groupby(['grouped_station', 'seasons', 'dayOfWeek', 'hour'], as_index=False).count()
trips.columns = ['grouped_station', 'seasons', 'dayOfWeek', 'hour', 'total_trips']
trips.total_trips.max()
```

```
import math 
# hourly average trips for given station (seasons - 3 months each, day of week - 4 weeks in a month(30/7), 4 years data(2016,2017,2018,2019), size of groups) 
def average_trips(num, group_id):
  return (int) (math.ceil(num / ((3 * 4 * 4) * group_length_dict.get(group_id))))
```

```
trips['total_trips'] = np.vectorize(average_trips)(trips['total_trips'], trips['grouped_station'])
```
